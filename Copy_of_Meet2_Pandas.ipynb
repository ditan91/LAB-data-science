### Pivot Table

Pivot table ini lebih mudah dimodifikasi dibanding group tadi tapi pivot table data nya harus satu nilai ( sama kaya kita make aggration tadi) dan menampilkan hanya apa yang kita definisikan di argumentnya tidak seperti group by

```python
#disi data yang dihasilkan auto mean nya dari saleprice
pd.pivot_table(df,index="HouseStyle",columns='YearBuilt',values='SalePrice',fill_value=0,aggfunc=numpy.mean)
```

```python
#seperti sebelumnya aggfunc bisa di isi fungsi buatan sendiri yang menghasilkan satu nilai

def std2(x):
    return numpy.std(x)*2

pd.pivot_table(df,index="HouseStyle",columns='YearBuilt',values='SalePrice',fill_value=0,aggfunc=std2)
```

```python
# bisa dibuat multiple column barisnya

pd.pivot_table(df,index=['HouseStyle','Foundation','Street'],columns=['YearBuilt'],values='SalePrice',fill_value=0,aggfunc=numpy.mean)
```

```python
# kita pake ikit aja biar gampang dibaca
pd.pivot_table(df,index="HouseStyle",columns='Street',values='SalePrice',fill_value=0,aggfunc=[numpy.mean,'count'])
```
